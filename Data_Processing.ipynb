```python
# import libraries 

import numpy as np
import pandas as pd
```

```python
# import dataset

ds=pd.read_csv("data.1.csv")
ds
```

```python
#install
!pip install scikit-learn
```

```python
#handlimg missing value(mean and scikit learn )
from sklearn.impute import SimpleImputer
si=SimpleImputer(missing_values = np.nan, strategy='mean')
ds[['Age','Salary']]=si.fit_transform(ds[['Age','Salary']])
```

```python
ds
```

```python
#encoding categorical column (OneHot Encoding)
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output = False)
encoded_data = ohe.fit_transform(ds[['Country']])
encoded_data
```

```python
ds.columns
```

```python
#make column header and make a datafrme for array
column_names = ohe.get_feature_names_out(['Country'])
encoded_df = pd.DataFrame(encoded_data, columns =column_names)
encoded_df
```

```python
column_names
```

```python
# Joing categorical column
ds = pd.concat([ds, encoded_df], axis = 1)
ds
```

```python
#droping categorical column
ds = ds.drop(columns = ['Country'])
ds
```

```python
# LabelEncoding 

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
ds['Purchased'] = le.fit_transform(ds['Purchased'])
```

```python
ds
```

```python
# Feature Scaling (Reduce Larger data values to its samller data values )

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
ds[['Age' , 'Salary']]= sc.fit_transform(ds[['Age', 'Salary']])
```

```python
# Divide dataset as Independent (input) and Dependent (output) varibles

x = ds[['Age','Salary','Country_France','Country_Germany','Country_Spain',]]
y = ds[['Purchased']]
```

```python
ds
```

```python
# Splitting dataset as Training and Testing

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=1)
```
